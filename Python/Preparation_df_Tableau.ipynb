```python
import pandas as pd
```

```python
df = pd.read_csv('df_clusters.csv')
```

```python
df.head()
```

```python
def replace_dot(df):
    for col in df.columns:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.replace('.', ',')
    return df
```

```python
df_new = replace_dot(df)
```

```python
df_new.head()
```

```python
df_new.to_csv('df_comma.csv')
```
